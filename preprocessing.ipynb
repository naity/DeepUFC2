```python
import pandas as pd
import numpy as np
```

```python
# read data for fights and fighters
bouts = pd.read_csv("ufc_bouts.csv", index_col = 0)
fighters = pd.read_csv("ufc_fighters.csv")
```

Next, We will perform a series of steps to clean up the datasets.

```python
bouts.head()
```

```python
fighters.head()
```

```python
# check if there are fighters with the same name
fighters[fighters.duplicated(subset="name", keep=False)]
```

```python
# note that we have several fighters who have the same names
# Fortunately, they belong to different weight classes
fighters.iloc[513, 1] = "Michael McDonald 205"
fighters.iloc[845, 1] = "Dong Hyun Kim 155"
fighters.iloc[954, 1] = "Tony Johnson 185"
fighters.iloc[1674, 1] = "Mike Davis 145"
```

```python
# use fighter names as index
fighters.drop("fighter_id", axis=1, inplace=True)
fighters.set_index("name", inplace=True)
```

```python
# Some fighters do not have statistics available, and we will remove those fighters.
fighers_clean = fighters.loc[~((fighters["SLpM"] == 0) &
                               (fighters["Str_Acc"] == "0%") & 
                               (fighters["SApM"] == 0) &
                               (fighters["Str_Def"] == "0%") &
                               (fighters["TD_Avg"] == 0) &
                               (fighters["TD_Acc"] == "0%") &
                               (fighters["TD_Def"] == "0%") &
                               (fighters["Sub_Avg"] == 0))].copy()                        
```

```python
print("{0} fighers in total, after clean up: {1} fighers".format(len(fighters), len(fighers_clean)))
```

Note:
* SLpM - Significant Strikes Landed per Minute
* Str_Acc - Significant Striking Accuracy
* SApM - Significant Strikes Absorbed per Minute
* Str_Def - Significant Strike Defence (the % of opponents strikes that did not land)
* TD_Avg - Average Takedowns Landed per 15 minutes
* TD_Acc - Takedown Accuracy
* TD_Def - Takedown Defense (the % of opponents TD attempts that did not land)
* Sub_Avg - Average Submissions Attempted per 15 minutes

```python
# add winning percentages
fighers_clean["win%"] = 100 * (fighers_clean["win"] / (fighers_clean["win"] +
                                                 fighers_clean["lose"] +
                                                 fighers_clean["draw"] +
                                                 fighers_clean["nc"]))
# change datatypes
percentages = ["Str_Acc", "Str_Def", "TD_Acc", "TD_Def"]
statistics = ["SLpM", "Str_Acc", "SApM", "Str_Def", "TD_Avg", "TD_Acc", "TD_Def", "Sub_Avg", "win%"]

fighers_clean.loc[:, percentages] = fighers_clean.loc[:, percentages].applymap(
    lambda x: x.replace("%", ""))

fighers_clean.loc[:, statistics] = fighers_clean.loc[:, statistics].astype(np.float32)
```

```python
fighers_clean.head()
```

```python
# use born year only
def get_year(dob):
    if pd.isna(dob):
        return "unk"
    return dob.split(", ")[-1]

fighers_clean["born_year"] = fighers_clean["dob"].map(lambda dob: get_year(dob))
```

```python
fighers_clean.head()
```

```python
# remove unused columns
fighers_clean.drop(["weight", "dob", "last_updated"], inplace=True, axis=1)
```

```python
fighers_clean.head()
```

Now, let's work on the fights dataset.

```python
bouts.head()
```

```python
bouts["result"].value_counts()
```

```python
# use matches that have a winner, ignore nc, draws
bouts = bouts.loc[bouts["result"] == "win"].copy()
```

```python
bouts["fight_year"] = bouts["date"].map(lambda d: d[:4])
bouts["method"] = bouts["method"].map(lambda m: m.split("-")[0])
```

```python
# remove unused columns
bouts.drop(["event_name", "date", "location", "attendance ", "result", "end_time"],
          axis=1, inplace=True)
```

```python
bouts.head()
```

```python
# winner is always fighter 1 
bouts["fighter1"].equals(bouts["winner"])
```

```python
# randomly swap fighter1 and fighter2 for half of the dataset to create "negative cases"
swap_indices = np.random.choice(len(bouts), size = len(bouts) // 2, replace = False)
bouts.iloc[swap_indices, [0, 1]] = bouts.iloc[swap_indices, [1, 0]].values
```

```python
bouts["winner"] = bouts["winner"] == bouts["fighter1"]
bouts["winner"] = bouts["winner"].astype(int)
bouts["winner"].value_counts()
```

```python
bouts.tail()
```

```python
bouts[bouts["fighter1"]=="Dong Hyun Kim"]
```

```python
# again, handel duplicate names
bouts_clean = bouts.copy()
for col in ["fighter1", "fighter2"]:
    bouts_clean.loc[(bouts_clean[col] == "Michael McDonald") &
                    (bouts_clean["weight_class"] == "Light Heavyweight"), col] = "Michael McDonald 205"
    
    bouts_clean.loc[(bouts_clean[col] == "Dong Hyun Kim") &
                    (bouts_clean["weight_class"] == "Lightweight"), col]= "Dong Hyun Kim 155"
    
    bouts_clean.loc[(bouts_clean[col] == "Tony Johnson") &
                    (bouts_clean["weight_class"] == "Middleweight"), col] = "Tony Johnson 185"
    
    bouts_clean.loc[(bouts_clean[col] == "Mike Davis") &
                    (bouts_clean["weight_class"] == "Featherweight"), col] = "Mike Davis 145"
```

```python
bouts_clean[bouts_clean["fighter1"] == "Dong Hyun Kim 155"]
```

Next, we need to fuse these two datasets together.

```python
# make sure all fighters are in the figher database
# otherwise remove the corresponding bout
all_fighter_names = fighers_clean.index.values.tolist()

bouts_clean = bouts_clean.loc[(bouts_clean["fighter1"].isin(all_fighter_names)) &
                              (bouts_clean["fighter2"].isin(all_fighter_names))]
```

```python
bouts_clean.reset_index(inplace=True, drop=True)
```

```python
bouts_clean.head()
```

```python
print("{0} bouts in total, after clean up: {1} bouts".format(len(bouts), len(bouts_clean)))
```

```python
fighter1_data = fighers_clean.loc[bouts_clean["fighter1"]]
fighter1_data = fighter1_data.add_suffix('_fighter1')
fighter2_data = fighers_clean.loc[bouts_clean["fighter2"]]
fighter2_data = fighter2_data.add_suffix('_fighter2')
```

```python
fighter1_data.head()
```

```python
fighter2_data.head()
```

```python
len(fighter1_data), len(fighter2_data)
```

```python
# now join the two datasets together
fighter1_data.reset_index(inplace=True, drop=True)
fighter2_data.reset_index(inplace=True, drop=True)
combined = pd.concat([bouts_clean, fighter1_data, fighter2_data], axis=1, sort=False)
```

```python
combined.to_csv("ufc_combined.csv")
```
